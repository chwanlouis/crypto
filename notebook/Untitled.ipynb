```python
# GLOBAL VARIABLES

BASE_REST_ENDPOINT = "https://api.cryptorank.io/v1/"
YOUR_API_KEY = ""
DATETIME_FORMAT = "%Y-%m-%dT%H:%M:%S.%f"  # ISO 8601 with microseconds yyyy-mm-ddThh:mm:ss.mmmZ
BROWSER_PARAMS = dict(
    origin='Chicago,IL',
    destination='Los+Angeles,CA',
    waypoints='Joplin,MO|Oklahoma+City,OK',
    sensor='false'
)
```

```python
import json
import requests


def requestCurrencyData(api_key):
    url = "%scurrencies?api_key=%s" % (BASE_REST_ENDPOINT, api_key)
    resp = requests.get(url=url, params=BROWSER_PARAMS)
    return resp.json()


currency_data = requestCurrencyData(YOUR_API_KEY)
```
